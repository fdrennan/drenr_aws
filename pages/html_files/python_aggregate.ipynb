```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
flights = pd.read_csv('http://drenr.com/datasets/flights.csv')
```

```python
flights.head(5)
```

```python
flights.drop(['Unnamed: 0'], axis=1)
```

```python
flights = flights.groupby(['year', 'month', 'day'])
```

```python
flights = flights.count()
```

```python
counts = flights.iloc[:,:1]
```

```python
x = np.linspace(1, 365, 365)
y = counts
```

```python
plt.figure(figsize=(15, 8))
plt.plot(x, y)
```
